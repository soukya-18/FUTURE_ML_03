```python
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity


resumes = [
    "Python Machine Learning Data Analysis",
    "Java Web Development HTML CSS",
    "Python Deep Learning AI",
    "Marketing Sales Communication"
]

job_description = "Python Machine Learning"


vectorizer = CountVectorizer()
X = vectorizer.fit_transform(resumes + [job_description])

similarity = cosine_similarity(X[-1], X[:-1])

print("Similarity Scores:")
print(similarity)


scores = similarity.flatten()
ranked = np.argsort(scores)[::-1]

print("\nRanking of candidates (best to worst):")
print(ranked)


print("\nBest Resume:")
for i in ranked:
    print(f"Resume: {resumes[i]} → Score: {scores[i]}")
```
